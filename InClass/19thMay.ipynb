```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download("stopwords")
nltk.download("punkt_tab")
#Sample Text:-
text3="This is a sample sentence, showing off the stop words filtration. "
text6="sadfaksdkjbhafdskbjhf dsfkhbjdFKBHJSDFAJHBK sdfjhbsdfkjvhdfs asdfsdafsd"
```

```python
token1=word_tokenize(text3)
token2=word_tokenize(text6)
```

```python
stop_word=set(stopwords.words("english"))
```

```python
filtered=[text3 for text3 in token1 if text3.lower() not in stop_word]
print(filtered)
```

```python
filtered3=[text6 for text6 in token2 if text6.lower() not in stop_word]
print(filtered3)
```

```python
from nltk.stem import PorterStemmer #Classic
from nltk.stem import SnowballStemmer #Its mid and for multiple languages
from nltk.stem import LancasterStemmer #Smarter But agressive
```

```python
words=["running", "ran", "runs", "easily", "fairly", "studies", "happily", "university", "relationally","delayed","tracking"]
wrods=["ich", "bin", "ein", "berliner"]
german="Sehr schlechter Kundenservice!!! Niemand antwortet auf meine Emails."
germanToken=word_tokenize(german)
porter=PorterStemmer()
snowballStemmer=SnowballStemmer("german")
lancaster=LancasterStemmer()
for i in words:
    print(f"Porter Stemmer: {porter.stem(i)}")
```

```python
# for i in wrods:
#     print(f"Snowball Stemmer: {snowballStemmer.stem(i)}")
for i in germanToken:
    print(f"Snowball Stemmer: {snowballStemmer.stem(i)}")
```

```python
for i in words:
    print(f"Lancaster Stemmer: {lancaster.stem(i)}")
```
