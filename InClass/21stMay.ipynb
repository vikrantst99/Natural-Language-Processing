## Dependency Parsing

```python
import spacy
nlp = spacy.load("en_core_web_sm")
text="Researchers analyze large datasets efficiently."
doc=nlp(text)
for i in doc:
    print(f"Token: {i.text:<15}"f"department: {i.dep_:<15}"f"Head: {i.head.text:<15}")
```

This basically shows everything.

Such as punctuations, subject atc.

```python
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Researchers analyze large datasets efficiently.")
displacy.render(doc, style="dep", jupyter=True) 
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Researchers analyze large datasets efficiently. asdsadf sdagfadsfasd sadfgasdfasdfsad")
displacy.render(doc, style="dep", jupyter=True) 
```

## NER (Named Entity Relationships)

It has challenges as well.

Such as Noun can be messed up.....

Such as a Dr. V or Proff. Dr. V are the same so.....

```python
text="Apple opened a new Vikrant in the year 2024"
doc=nlp(text)
for i in doc.ents:
    print(i.text,i.label_)
```

```python
for i in doc.ents:
    print(f"Entity: {i.text:<15}"f"Label: {i.label_:<15}"f"Explanations: {spacy.explain(i.label_):<15}")
```

```python
nlp = spacy.load("en_core_web_sm")
displacy.render(doc, style="dep", jupyter=True) 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents=["I love Natural Language Processing","I love Machine Learning","Machine Learning is powerful"]
verctorizer=TfidfVectorizer()
tfidf_matrix=verctorizer.fit_transform(documents)
vocab=verctorizer.get_feature_names_out()
print(vocab)
print(tfidf_matrix.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
cosineSimMat=cosine_similarity(tfidf_matrix)
print(cosineSimMat)
csoScore=cosineSimMat[0][1]
print("")
print("Cosine Similarity Score between document 1 and document 2 is: ")
print(csoScore)
# data=pd.DataFrame(cosineSimMat,index=[csoScore],columns=documents)
# data
```
