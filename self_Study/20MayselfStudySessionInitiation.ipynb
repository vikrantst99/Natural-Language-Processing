A range can be defined using a dash '-' in square brackets '[]' 

| Pattern | Meaning |
| :--- | :--- |
| `[A-Z]` | one uppercase English letter |
| `[a-z]` | one lowercase English letter |

This symbol inside the square brackets means "MATCH ANYTHING EXCEPT these characters.


| Pattern | Meaning |
| :--- | :--- |
| `[^0-9]` | not a digit |
| `[^a-z]` | not a lowercase letter |
| `[^Ss]` | not S and not s |


EXAMPLES:- 

```python
LowerCases=re.findall(r"[^A-Z]",textLong)
noDigits=re.findall(r"[^0-9]",textLong)
capsOnly=re.findall(r"[^a-z]",textLong)
print(f"Lowercase letters are: \n{LowerCases}\n\n")
print(f"Non-digit characters are: \n{noDigits}\n\n")
print(f"Uppercase letters are: \n{capsOnly}")
```

# Quantifiers : HOW MANY TIMES.

Quantifiers control repetition.

# A. '+' means one or more than one times:-

If I use r"\d+" , for a text set, Outputs can be 1,20,30, 300, 9000, etc.

```python
text = "I have 2 exams and 15 assignments."
result=re.findall(r"\d+",text)
result2=re.findall(r"\w+",text)
result3=re.findall(r"\s+",text)
print(result)
print(result2)
print(result3)
```

# B. * means zero or more 

```python
text = "coluor colour colouur kolor color colon colkr"
result = re.findall(r"colo*u*r", text)
print(result)
```

This can match multiple spellings because o* and u* allow flexible repetition.

# C. ? means zero times or one.

In the aforementioned output, I understood that:-

# D. {n} means eaxctly n number of times.

Example:-

```python
text="My Name Is Vikrant"
text2="2024, 192, 1987, 23"
result=re.findall(r"\w{7}",text)
result2=re.findall(r"\d{4}",text2)
print(result)
print(result2)
```

# E. {n,} AND \ {n,m} means atleast n times 

1. r"\d{2,} means Match 2 or more digits. 
2. r"\d{2,10} means match 2 to 10 digits.

Examples:-

```python
text1="22, 333, 4444, 55555, 666666, 7777777, 88888888, 999999999, 1010101010"
#Normal {3} means exactly 3 occurrences of the preceding element, so it will match any sequence of exactly 3 digits in the text.
s=re.findall(r"\d{3}",text1)
# {3,} means 3 or more occurrences of the preceding element, so it will match any sequence of 3 or more digits in the text.
result=re.findall(r"\d{3,}",text1)
# {3,6} means between 3 and 6 occurrences (inclusive), so it will match any sequence of 3 to 6 digits in the text.
result2=re.findall(r"\d{3,6}",text1)
print(s)
print(result)
print(result2)
```

# F. DOT '.' means :- Any single character except a new line.

Examples:-

```python
text = "begin begun began beginning begunning beg begginium"
result = re.findall(r"beg.n", text)
print(result)
# In the output, you will see that "begun", "began", and "begginium" 
# are matched because they all start with "beg", followed by any single character 
# (which can be "u", "a", or "g"), and end with "n". The other words like "begin", 
# "beginning", and "begunning" do not match the pattern because they do not have exactly 
# one character between "beg" and "n".
#https://prnt.sc/hzNjxIbRZdXO   Check the link.
```

# G. Anchors: ^ and $ :-

They do not match characters instead they match positions.

# 1. ^ means start of a String:-
r"^Hello" Matches :- Hello Vikrant and will not match :- Hi Hello Vikrant.
# 2. $ means end of String :- 
r"Germany$" Matches :- I live in Germany and will not match :- Germany is nice

```python
text= "HELLO!!! I am learning NLP. HELLO ISNT IT. HELLO HOW ARE YOU? HI HELLO. Hello"
text2="I live in Germany"
print(re.findall(r"^HELLO",text)) 
#^HELLO matches "HELLO" at the beginning of the string. 
# The ^ symbol indicates that the match should occur 
# at the start of the string. In this case, since "HELLO" 
# is indeed at the beginning of the text variable, it will 
# be matched and returned in the output. If you were to search 
# for a pattern that does not exist at the beginning of the string, 
# it would not be matched.
print(re.findall(r"Germany$",text2)) #Germany$ matches "Germany" at the end of the string.
```

# J. Groups '()' :- It groups parts of a pattern.

Check the screenshot:- 



```python
gmail="vikrantst99@gmail.com"
print(re.findall(r"(\w+)@(\w+)",gmail)) 
#The pattern r"(\w+)@(\w+)" is designed to match email addresses in the format of "username@domain".
# Here's a breakdown of the pattern:
# (\w+): This captures one or more word characters (letters, digits, or underscores) for the username.
# @: This matches the literal "@" symbol.
# (\w+): This captures one or more word characters for the domain.
#OR:-
print(re.findall(r"@(\w+)",gmail)) 
#This pattern r"@(\w+)" is designed to match the domain part of an email address that follows the "@" symbol.
```

| Group | Captures |
| :--- | :--- |
| `(\w+) before @` | username |
| `(\w+) after @` | domain name |

# K. Escaping Special Characters.

Some characters have special regex meanings:


. + * ? ^ $ ( ) [ ] { } | \

| You want to match | Regex | Meaning | Examples |
| :--- | :--- | :--- | :--- |
| `.` | `\.` | Matches a literal period/dot | Matching file extensions like `\.txt` or `\.jpg` |
| `+` | `\+` | Matches a literal plus sign | Matching a blood type like `A\+` or math expressions like `5\+5` |
| `?` | `\?` | Matches a literal question mark | Matching questions like `Why\?` or `How\?` |
| `(` | `\(` | Matches a literal opening parenthesis | Matching area codes in phone numbers like `\(\d{3}\)` |
| `)` | `\)` | Matches a literal closing parenthesis | Closing the area code example above: `\(\d{3}\)` |
| `[` | `\[` | Matches a literal opening square bracket | Matching shortcode patterns or array syntax like `data\[0\]` |

# ⚠️ Important Warning: The Limits of Regex

> **Crucial Takeaway:** Regular expressions are incredibly powerful, but they are built for syntactic patterns, not semantic understanding. Knowing when *not* to use regex is just as important as knowing how to write it.

---

### 🎯 Where Regex Excels: Fixed Patterns
Regex is your go-to tool when the data follows a predictable, structural rule. It is highly efficient for extracting:
* 📧 **Emails**
* 📱 **Phone numbers**
* 📅 **Dates**
* 🏷️ **Hashtags**
* 🌐 **URLs**
* 🆔 **IDs**
* 🧾 **Invoice numbers**
* 🔍 **Simple keyword rules**

---

### ❌ Where Regex Fails: Deep Meaning
Regex matches characters, not context. It completely falls apart when trying to decode the nuances of human language, such as:
1.  🎭 **Sarcasm** *(Detecting if "Oh, great" means good or bad)*
2.  ❤️ **Emotion & Sentiment** *(Understanding the mood behind a paragraph)*
3.  🧠 **Sentence meaning** *(Grasping the core message of varied phrasing)*
4.  🔄 **Pronoun resolution** *(Figuring out who "he" or "it" refers to in a long text)*
5.  🤷 **Ambiguous names** *(Distinguishing between Apple the company and apple the fruit)*
6.  🗣️ **Complex grammar** *(Handling shifting sentence structures)*

---

### 🚀 Moving Beyond: The NLP Evolution
Because human language is messy, Natural Language Processing (NLP) eventually transitions away from hardcoded regex patterns and moves toward advanced linguistic models:

* **Tokenization:** Breaking text into smart semantic pieces.
* **POS Tagging:** Labeling parts of speech (nouns, verbs, adjectives) dynamically.
* **NER (Named Entity Recognition):** Identifying real-world entities like people, places, or dates regardless of how they are written.
* **Embeddings:** Converting words into mathematical vectors that capture their literal meaning and relationships.
* **Transformers:** Utilizing massive deep learning architectures (like BERT or GPT) to understand full context.

> 🛠️ **The Golden Rule:** While NLP handles the heavy cognitive lifting, **regex remains a core foundational tool** in real-world NLP pipelines for initial data cleaning, preprocessing, and filtering.
