## # 1. Core idea

| Concept | Meaning | Example |
| :--- | :--- | :--- |
| **Regex** | A pattern used to find text | find numbers, emails, dates |
| **Pattern** | The rule we give to Python | `r"\d+"` |
| **Match** | Text that fits the pattern | `25` matches `\d+` |
| **Raw string** | Python regex string written with `r` | `r"\d+"` |

## # 2. Python regex functions

| Function | Use | Example |
| :--- | :--- | :--- |
| **re.findall()** | find all matches | `re.findall(r"\d+", text)` |
| **re.search()** | find first match anywhere | `re.search(r"NLP", text)` |
| **re.match()** | match only at beginning | `re.match(r"Hello", text)` |
| **re.sub()** | replace matches | `re.sub(r"\s+", " ", text)` |
| **re.split()** | split using pattern | `re.split(r",", text)` |
| **re.finditer()** | loop through matches (useful for positions) | `re.finditer(r"\d+", text)` |
| **groups `()`** | capture only part of match | `r"delay (\d+)"` |

## # 3. Character patterns

| Regex | Meaning | Example match |
| :--- | :--- | :--- |
| `[abc]` | a, b, or c | `a` |
| `[A-Z]` | uppercase letter | `N` |
| `[a-z]` | lowercase letter | `p` |
| `[0-9]` | digit | `7` |
| `[A-Za-z]` | any English letter | `G` |
| `[A-Za-z0-9]` | letter or digit | `A`, `8` |
| `[^0-9]` | not a digit | `A`, ` `, `!` |
| `[^A-Za-z0-9\s]` | punctuation/symbols | `!`, `?`, `-` |

## # 4. Shortcut patterns

| Regex | Meaning | Similar to |
| :--- | :--- | :--- |
| `\d` | digit | `[0-9]` |
| `\D` | not digit | `[^0-9]` |
| `\w` | word character: letter, digit, `_` | `[A-Za-z0-9_]` |
| `\W` | not word character (punctuation, spaces) | `[^A-Za-z0-9_]` |
| `\s` | whitespace (space, tab, newline) | `[ \t\n\r\f\v]` |
| `\S` | not whitespace (letters, digits, symbols) | `[^\s]` |

## # 5. Repetition symbols

| Regex | Meaning | Example |
| :--- | :--- | :--- |
| `+` | one or more | `\d+` finds `25` |
| `*` | zero or more | `a*` can match `""`, `a`, `aaa` |
| `?` | zero or one | `walks?` finds `walk`, `walks` |
| `{2}` | exactly 2 times | `\d{2}` finds `23` |
| `{2,}` | 2 or more times | `[A-Z]{2,}` finds `NLP` |
| `{2,4}` | 2 to 4 times | `\d{2,4}` finds `572`, `2026` |

## # 6. Position symbols

| Regex | Meaning | Example |
| :--- | :--- | :--- |
| `^` | start of text/line | `^Hello` matches "Hello" only if it's at the very beginning. |
| `$` | end of text/line | `Germany$` matches "Germany" only if it's at the very end. |
| `\b` | word boundary | `\bcat\b` matches the exact word "cat" (ignores "cattle" or "bobcat"). |
| `.` | any character except newline | `beg.n` matches `begin`, `begun`, `began`, or `beg-n`. |

## # 7. Logic and grouping

| Regex | Meaning | Example |
| :--- | :--- | :--- |
| `\|` | OR operator | `cat\|dog` matches either "cat" or "dog" |
| `( )` | capture group | `r"delay (\d+)"` isolates just the digits after "delay" |
| `(?: )` | non-capturing group | `(?:https?://)` groups options together without saving them as a standalone result |
| `\.` | actual dot | `\.` matches a literal period `.` instead of acting as a wildcard |
| `\+` | actual plus | `\+` matches a literal plus sign `+` instead of meaning "one or more" |
| `\-` | actual hyphen | `\-` matches a literal dash/hyphen `-` (essential when inside character classes like `[a\-z]`) |

## # 8. Common NLP extraction patterns

| Goal | Regex Pattern | Example Match |
| :--- | :--- | :--- |
| **All numbers** | `r"\d+"` | `572`, `25`, `2026` |
| **Time like 18:45** | `r"\b\d{2}:\d{2}\b"` | `18:45` |
| **German date like 23.06.2026** | `r"\b\d{2}\.\d{2}\.\d{4}\b"` | `23.06.2026` |
| **Email** | `r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"` | `support@db.de` |
| **Hashtag** | `r"#[A-Za-z0-9_]+"` | `#NLP`, `#Python3` |
| **Hashtag without #** | `r"#([A-Za-z0-9_]+)"` | Extracts `NLP` from `#NLP` |
| **Twitter handle from URL** | `r"https://twitter\.com/([A-Za-z0-9_]+)"` | Extracts `iamvikrant` from the URL |
| **URL** | `r"https?://\S+"` | `https://google.com` or `http://db.de` |
| **Capital code like NLP** | `r"\b[A-Z]{2,}\b"` | `NLP`, `ICE`, `VSCODE` |
| **Room like A-304** | `r"\b[A-Z]-\d{3}\b"` | `A-304`, `B-102` |
| **Repo name with hyphens** | `r"[A-Za-z-]+"` | `scikit-learn`, `regex-practice` |

## # 9. Cleaning patterns

| Cleaning Task | Regex / Python Solution | Impact (Before вћЎпёЏ After) |
| :--- | :--- | :--- |
| **Remove punctuation safely** | `re.sub(r"[^A-Za-z0-9\s]", " ", text)` | `"Hello!!!"` вћЎпёЏ `"Hello   "` |
| **Collapse many spaces** | `re.sub(r"\s+", " ", text)` | `"Hello   World"` вћЎпёЏ `"Hello World"` |
| **Normalize to lowercase** | `text.lower()` | `"NLP"` вћЎпёЏ `"nlp"` |
| **Remove start/end spaces** | `text.strip()` | `"  nlp  "` вћЎпёЏ `"nlp"` |

## # Example of full cleaning flow:- (Very Short):- 

```python
import re
text = "NLP!!!     is amazing... Regex??? is useful; Python---is powerful :)"
cleaned = re.sub(r"[^A-Za-z0-9\s]", " ", text)
cleaned = re.sub(r"\s+", " ", cleaned)
cleaned = cleaned.lower()
cleaned = cleaned.strip()
print(cleaned)
```
