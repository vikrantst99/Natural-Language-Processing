# 🛠️ Practice Exercise 1: Train Delay

### 📝 Practice Text
> "Today my train ICE 572 is delayed by 25 minutes.  
> Platform changed from 4 to 7.  
> New departure time is 18:45.  
> Contact support at bahn_support@db.de."

---

### 🎯 Tasks

| No. | Extract Goal | Target Data | Suggested Regex Pattern |
| :--- | :--- | :--- | :--- |
| **1** | All numbers | `572`, `25`, `4`, `7`, `18`, `45` | |
| **2** | Train name | `ICE 572` | |
| **3** | Delay minutes | `25` | |
| **4** | Platform numbers | `4` and `7` | |
| **5** | Time | `18:45` | |
| **6** | Email address | `bahn_support@db.de` | |

# Task 1:-

```python
text = """Today my train ICE 572 is delayed by 25 minutes.
Platform changed from 4 to 7.
New departure time is 18:45.
Contact support at bahn_support@db.de.
"""
```

```python
import re
```

```python
print(re.findall(r"\d+", text))
```

# Task 2:- 

| Regex part | Meaning |
| :--- | :--- |
| `\b` | word boundary |
| `[A-Z]` | one capital letter |
| `{2,}` | two or more times |
| `\s` | one space |
| `\d+` | one or more digits |
| `\b` | word boundary |

```python
print(re.findall(r"\b[A-Z]{2,}\s\d+\b", text))
```

# Task 3:- 

```python
delay = re.findall(r"delayed by (\d+) minutes", text)
print(delay)
```

# Task 4:- 

```python
platforms = re.findall(r"Platform changed from (\d+) to (\d+)", text)
print(platforms)
```

# Task 5:- 

```python
time = re.findall(r"\b\d{2}:\d{2}\b", text)
print(time)
```

EXPLANATION:- 

| Regex part | Meaning |
| :--- | :--- |
| `\d{2}` | exactly 2 digits |
| `:` | literal colon |
| `\d{2}` | exactly 2 digits |

# Task 6:- 

```python
email = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
print(email)
```
