# 🧹 Practice Exercise 2: Cleaning Messy Text

### 📝 Practice Text
> "NLP!!!     is amazing... Regex??? is useful; Python---is powerful :)"

---

### 🎯 Cleaning Tasks

| No. | Step | Target Action | Logic / Suggested Tool |
| :--- | :--- | :--- | :--- |
| **1** | Remove Punctuation | Strip out `!!!`, `...`, `???`, `;`, `---`, and `:)` | Use `re.sub()` with a character class like `[^A-Za-z0-9\s]` to replace punctuation with an empty string. |
| **2** | Fix Whitespace | Reduce multiple spaces down to a single space | Use `re.sub()` to catch `\s+` (one or more spaces) and replace them with a single `" "`. |
| **3** | Normalize Case | Lowercase every letter | Use Python's built-in `.lower()` string method. |

---

### 🏁 Expected Output
Once you apply your steps, your final string should look exactly like this:
`nlp is amazing regex is useful python is powerful`

```python
text = "NLP!!!     is amazing... Regex??? is useful; Python---is powerful :)"
```

1. Step 1:- We replace the punctuations with spaces.

```python
import re
cleaned=re.sub(r"[^A-Za-z0-9\s]", " ", text)
print(cleaned)
```

2. Step 2:- Replace many spaces with just one.

```python
cleaned=re.sub(r"\s+", " ", cleaned)
print(cleaned)
```

3. Step 3:- Lower Case

```python
cleaned=cleaned.lower()
```

4. Step 4:- Remove beginning and end spaces

```python
cleaned=cleaned.strip()
print(cleaned)
```

| Goal | Regex Pattern | Reasons (Why we do it) | Examples |
| :--- | :--- | :--- | :--- |
| **Numbers** | `r"\d+"` | To extract any sequence of digits from text, regardless of how long the number is. | Extracts `572` and `25` from train delays. |
| **Time like 18:45** | `r"\b\d{2}:\d{2}\b"` | To find timestamps while using `\b` to prevent matching parts of longer numbers (like serial codes). | Extracts `18:45` but ignores `118:452`. |
| **German date** | `r"\b\d{2}\.\d{2}\.\d{4}\b"` | To capture standard European/German date formats (DD.MM.YYYY), escaping the periods so they match literal dots. | Extracts `20.05.2026`. |
| **Email** | `r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"` | To capture complex text strings structured as a username, an `@` symbol, a domain, and a valid extension. | Extracts `bahn_support@db.de` or `vikrant@gmail.com`. |
| **Hashtag** | `r"#[A-Za-z0-9_]+"` | To extract full trending topics or tags, including the hash symbol itself, from social media text. | Extracts `#NLP` or `#Python_Programming`. |
| **Hashtag without #** | `r"#([A-Za-z0-9_]+)"` | To isolate just the *text value* of a hashtag. The parentheses `()` create a capture group, allowing you to ignore the `#` in your final output. | From `#Python`, it isolates and extracts just `Python`. |
| **URL** | `r"https?://\S+"` | To grab web links. The `s?` makes the 's' optional (matching `http` or `https`), and `\S+` grabs everything until the next space. | Extracts `https://google.com`. |
| **Multiple spaces** | `r"\s+"` | To target messy spacing (2 or more spaces, tabs, or newlines) so you can collapse them into a single clean space. | Finds the massive gaps in `"NLP!!!     is amazing"`. |
| **Remove punctuation safely** | `r"[^A-Za-z0-9\s]"` | The `^` inside the brackets means "NOT". It targets everything that isn't a letter, number, or space, allowing you to strip punctuation without breaking words. | Targets and removes `!!!`, `???`, and `---`. |
| **Capital code like NLP** | `r"\b[A-Z]{2,}\b"` | To find acronyms, stock tickers, or capitalized codes that are at least 2 letters long, ignoring lowercase text. | Extracts `NLP` and `ICE` but ignores `Today`. |
