## Term Frequency 

Means how many times a word appears in 'one' document.

## 🔢 Term Frequency Concepts Breakdown



| Term | Meaning |
| :--- | :--- |
| **Term** | A word or token that we are tracking and counting. |
| **Frequency** | The total number of times any specific event or thing happens. |
| **Term Frequency (TF)** | The exact number of times a specific word appears inside a single text document. |

This is directly connected to Bag of Words and TF-IDF, where text is turned into word-count-based features.
